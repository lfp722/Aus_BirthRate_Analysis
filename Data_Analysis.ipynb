```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Apple Gothic'  #font
matplotlib.rcParams['font.size'] = 15  # font size
```

```python
fig, ax1 = plt.subplots(figsize=(15,10))
ax1.plot(df.index, df['Number of Baby'], color='#ff812d')
ax2 = ax1.twinx()
ax2.plot(df.index, df['Fr'])
```

```python
fig, ax1 = plt.subplots(figsize=(10,7))
ax1.bar(df.index, df['Number of Baby'], color='#ff812d')
ax1.set_ylabel('Number of Baby, 1000')
ax1.set_ylim(250, 400)

ax2 = ax1.twinx()
ax2.set_ylabel('Fertilization Rate')
ax2.set_yticks([0, 1])
ax2.plot(df.index, df['Fr'], marker='o', ms=15, lw=5)
```
